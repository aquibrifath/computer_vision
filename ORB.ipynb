```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img1 = cv2.imread('/home/aquirah/Downloads/IMG_0644.JPG',cv2.IMREAD_GRAYSCALE) 


scale_percent = 20 # percent of original size
width = int(img1.shape[1] * scale_percent / 100)
height = int(img1.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
resized1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)

  
cv2.imshow("Resized image", resized1)
cv2.waitKey(0)
cv2.destroyAllWindows()



```

```python
# Initialize the ORB detector with custom parameters
orb = cv2.ORB_create(
    nfeatures=500,      # Maximum number of keypoints to be detected.
    scaleFactor=2,    # Pyramid scale factor between the levels.
    nlevels=10,          # Number of pyramid levels.
    edgeThreshold=15,   # Threshold to filter out keypoints on edges.
    patchSize=150       # Size of the patch used by the oriented BRIEF descriptor.
)

# Detect keypoints
keypoints = orb.detect(resized1, None)
keypoints = sorted(keypoints, key=lambda x: x.response, reverse=True)



# Draw keypoints on the image
image_with_keypoints = cv2.drawKeypoints(resized1, keypoints[:100], None)

# Display the image with keypoints
cv2.imshow('ORB Keypoints', image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
