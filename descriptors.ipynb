```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
print("Hello world")
```

```python
img = cv2.imread('/home/aquirah/Pictures/Aquib_image.jpg') 
```

```python
cv2.imshow('Original', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
scale_percent = 25 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray scale image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
cv2.imshow("Gray scale image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(121),plt.imshow(resized),plt.title('Original')
plt.subplot(122),plt.imshow(blurred),plt.title('Blurred')
plt.show()
```

```python
wide = cv2.Canny(blurred, 100, 100)
```

```python
cv2.imshow("canny ", wide)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.subplot(121),plt.imshow(gray),plt.title('Original')
plt.subplot(122),plt.imshow(wide),plt.title('Blurred')
plt.show()
```

```python
yo=gray
yo = np.float32(yo)
dst = cv2.cornerHarris(yo, 2,5,0.07)

# Dilate the corner points to enhance visualization
dst = cv2.dilate(dst, None)


# Create a copy of the original image
image_copy = np.copy(resized)

# Mark the corners on the image
image_copy[dst > 0.01 * dst.max()] = [0, 0, 255]  # Mark corners in red (BGR color space)

cv2.imshow('Corners Detected', image_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
operatedImage = gray
  
operatedImage = np.float32(operatedImage)

dest = cv2.cornerHarris(operatedImage, 2, 5, 0.07)
  
dest = cv2.dilate(dest, None)

gray[dest > 0.01 * dest.max()]=[0, 0, 255]
  
cv2.imshow('Image with Borders', resized)
  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = sift.detectAndCompute(gray, None)

# Draw keypoints on the image
image_with_keypoints = cv2.drawKeypoints(resized, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with keypoints
cv2.imshow('Image with SIFT Keypoints', image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
