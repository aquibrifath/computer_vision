```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('/home/aquirah/Downloads/IMG_0643.JPG',cv2.IMREAD_GRAYSCALE) 
img2 = cv2.imread('/home/aquirah/Downloads/IMG_0644.JPG',cv2.IMREAD_GRAYSCALE) 

```

```python
scale_percent = 20 # percent of original size
width = int(img1.shape[1] * scale_percent / 100)
height = int(img1.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
resized1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
resized2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)

  
cv2.imshow("Resized image", resized2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# image_copy1 = cv2.cvtColor(resized1, cv2.COLOR_BGR2RGB)
# image_copy2 = cv2.cvtColor(resized2, cv2.COLOR_BGR2RGB)

image_copy1=resized1
image_copy2=resized2
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,12))

axs[0].imshow(image_copy1)
axs[0].set_title('image1')

axs[1].imshow(image_copy2)
axs[1].set_title('image2')

plt.show()
```

```python
sift = cv2.SIFT_create()


# Detect keypoints and compute descriptors
keypoints1, descriptors1 = sift.detectAndCompute(image_copy1, None)
keypoints2, descriptors2 = sift.detectAndCompute(image_copy2, None)


# Draw keypoints on the image
image_with_keypoints1 = cv2.drawKeypoints(image_copy1, keypoints1, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
image_with_keypoints2 = cv2.drawKeypoints(image_copy2, keypoints2, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with keypoints
cv2.imshow('Image with SIFT Keypoints', image_with_keypoints1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,12))

axs[0].imshow(image_with_keypoints1)
axs[0].set_title('image1')

axs[1].imshow(image_with_keypoints2)
axs[1].set_title('image2')

plt.show()
```

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

# Match the descriptors
matches = bf.match(descriptors1, descriptors2)

# Sort the matches by distance
matches = sorted(matches, key=lambda x: x.distance)

# Draw the top 10 matches
matched_image = cv2.drawMatches(image_copy1, keypoints1, image_copy2, keypoints2, matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Display the matched image
cv2.imshow('KeyPoint Matching', matched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Using KNN



# Initialize the BFMatcher
bf1 = cv2.BFMatcher()

# Perform kNN matching
matches = bf1.knnMatch(descriptors1, descriptors2, k=2)

# Apply Lowe's ratio test to filter good matches
good_matches = []
for m, n in matches:
    if m.distance < 0.45 * n.distance:
        good_matches.append(m)

# Draw the good matches
matched_image = cv2.drawMatches(image_copy1, keypoints1, image_copy2, keypoints2, good_matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Display the matched image
cv2.imshow('KeyPoint Matching', matched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
